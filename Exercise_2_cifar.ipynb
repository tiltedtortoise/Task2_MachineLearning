#### Loading data from the file system and generating training and test sets

To load the images from the file system we need to walk through the folders and add the images (normalized) to an array as well as the corresponding labels. Also, for the machine learning algorithms to work with these arrays we need to convert them to numpy arrays

---

#### Task 03 - Load images as features and labels  

##### Finish the function to load the images and their corresponding labels and return them as numpy arrays
* loop through the directory with os.walk https://www.geeksforgeeks.org/os-walk-python/
* create a filepath for each file
* and load each file with mpimg.imread https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imread.html
* normalize each image https://medium.com/analytics-vidhya/a-tip-a-day-python-tip-8-why-should-we-normalize-image-pixel-values-or-divide-by-255-4608ac5cd26a
* do not forget the labels
* return them as a numpy array

```python
def load_images_as_features_and_labels(target_dir, class_name_dictionary):
    """
    returns 
    X -- images as numpy array
    y -- labels as numpy array
    """
    X = []
    y = []

    # walk through target_dir, visit each class folder and its files
    for dirpath, dirnames, filenames in os.walk(target_dir):
        folder_name = os.path.basename(dirpath)

        # only handle folders that are actual class names
        if folder_name in class_name_dictionary:
            label = class_name_dictionary[folder_name]

            for fname in filenames:
                # only accept common image extensions
                fn = fname.lower()
                if fn.endswith((".png", ".jpg", ".jpeg")):
                    fpath = os.path.join(dirpath, fname)

                    # read image
                    img = mpimg.imread(fpath)

                    # ensure float32 and normalize to [0,1] if needed
                    img = np.array(img, dtype=np.float32)
                    if img.max() > 1.0:
                        img = img / 255.0

                    X.append(img)
                    y.append(label)

    # convert to numpy arrays
    X = np.stack(X, axis=0)        # shape: (N, 32, 32, 3)
    y = np.array(y, dtype=np.int64)  # shape: (N,)
    # TODO: Add your code here
    return X, y
```

With help of the function above we can now load the images and there labels as a training and test set (this could take a while, depending on your hardware!)

```python
X_train, y_train = load_images_as_features_and_labels('data/CIFAR-10-images/train/', class_names_dict)
X_test, y_test = load_images_as_features_and_labels('data/CIFAR-10-images/test/', class_names_dict)

X_train, y_train, X_test, y_test
```

#### TASK 07 - Load your datasets for training and testing the DL model

```python
#This part is responsible for creating the train_data and test_data, i.e., loading and organizing the image data in a way that is usable by TensorFlow for training a neural network.
import tensorflow as tf

train_dir = "data/CIFAR-10-images/train/"
test_dir = "data/CIFAR-10-images/test/"

BATCH_SIZE = 64
IMG_SIZE = (32, 32) # CIFAR-10 images are 32x32

# TODO load the training set
train_data = tf.keras.utils.image_dataset_from_directory(
    directory=train_dir,
    image_size=IMG_SIZE,
    label_mode="int",        # Labels are integers (0-9)
    batch_size=BATCH_SIZE,
    shuffle=True             # IMPORTANT: Shuffle training data!
)

# TODO load the testing set
test_data = tf.keras.utils.image_dataset_from_directory(
    directory=test_dir,
    image_size=IMG_SIZE,
    label_mode="int",        # Labels are integers (0-9)
    batch_size=BATCH_SIZE,
    shuffle=False            # No need to shuffle test data
)
def one_hot_encode_labels(image, label):
    # num_classes is 10
    label = tf.one_hot(label, depth=10)
    return image, label


print(f"Train Dataset: {train_data}")
print(f"Test Dataset:  {test_data}")
```

#### Task 08 - Build a sequential model with preprocessing layers
##### Use keras sequential API ( https://www.tensorflow.org/api_docs/python/tf/keras/Sequential) to stack preprocessing layers. Do not forget to normalize/rescale your data!

```python
# create a sequential model which will be added as a layer in the model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Rescaling
# Note: In newer TensorFlow versions (2.12+), preprocessing layers are moved
# from experimental to the standard tf.keras.layers namespace.
#from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomZoom

# TASK 08 - Build a sequential model with preprocessing layers
# We will create a preprocessing pipeline that rescales the data and optionally performs augmentation.

data_augmentation = Sequential(
    [
        # 1. Rescaling/Normalization (MANDATORY)
        # Rescales pixel values from 0-255 to 0-1. This is crucial for model stability.
        Rescaling(1./255),

        # 2. Data Augmentation (OPTIONAL but RECOMMENDED for better generalization)
        # These layers perform random transformations on the fly during training.
        RandomFlip("horizontal"),
        RandomRotation(0.2), # Rotate by up to 20%
        RandomZoom(0.2),    # Zoom in/out by up to 20%
    ],
    name="data_augmentation_and_normalization"
)
```

---

Next we build our CNN model, this time with help of the functional API.  
It is the same model as before, but with addition of our `data_augmentation` layer

---

#### Task 09 - Build a model with the functional API
##### Use keras functional API (https://www.tensorflow.org/guide/keras/functional) to build your CNN do not forget to include your data_augmentation layer
You can use layers like: 
* Conv2D, 
* Dense,
* BatchNormalization, 
* MaxPooling,
* Dropout,
* Flatten (for input into Dense)
* ....

<p><strong style="color:red">Attention: </strong>Experiment with different layers and number of layers until you reach at least 70% accuracy</p>

```python
# build model with functional api
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout

# Assume num_classes = 10 is already defined from the previous step

# Define a robust CNN architecture using the Functional API
# tf.keras.layers.Input: Defines the input tensor's shape (32x32 pixels, 3 color channels).
# This is the starting point for the Functional API model.
inputs = tf.keras.layers.Input(shape=(32, 32, 3), name='input_layer')

# 1. Preprocessing and Augmentation Pipeline
# data_augmentation(inputs): Applies the sequence of preprocessing layers (e.g., Rescaling, RandomFlip)
# to the input tensor, making the model more robust to data variations.
x = data_augmentation(inputs) # Apply the sequential data_augmentation model first

# 2. Convolutional Block 1 (High capacity, uses BatchNormalization for stability)
# Conv2D: Learns 32 spatial features (kernels) from the input using a 3x3 window. 'relu' introduces non-linearity.
x = Conv2D(32, 3, padding='same', activation='relu', name='conv_1')(x)
# BatchNormalization: Normalizes the activations of the previous layer, which stabilizes and speeds up training.
x = BatchNormalization(name='bn_1')(x)
# Conv2D: A second convolutional layer to extract deeper features.
x = Conv2D(32, 3, padding='same', activation='relu', name='conv_2')(x)
# MaxPooling2D: Downsamples the spatial dimensions (halves height/width) to reduce parameters and achieve translation invariance.
x = MaxPooling2D(pool_size=(2, 2), name='maxpool_1')(x)
# Dropout: Randomly sets 25% of input units to 0 during training to prevent overfitting.
x = Dropout(0.25, name='dropout_1')(x)

# 3. Convolutional Block 2
# Conv2D: Increases the number of filters to 64 to learn more complex features.
x = Conv2D(64, 3, padding='same', activation='relu', name='conv_3')(x)
# BatchNormalization: Normalizes activations.
x = BatchNormalization(name='bn_2')(x)
# Conv2D: Second 64-filter layer.
x = Conv2D(64, 3, padding='same', activation='relu', name='conv_4')(x)
# MaxPooling2D: Further downsamples the features.
x = MaxPooling2D(pool_size=(2, 2), name='maxpool_2')(x)
# Dropout: Randomly drops 25% of units.
x = Dropout(0.25, name='dropout_2')(x)

# 4. Convolutional Block 3
# Conv2D: Increases the number of filters to 128.
x = Conv2D(128, 3, padding='same', activation='relu', name='conv_5')(x)
# BatchNormalization: Normalizes activations.
x = BatchNormalization(name='bn_3')(x)
# MaxPooling2D: Final downsampling of feature maps.
x = MaxPooling2D(pool_size=(2, 2), name='maxpool_3')(x)
# Dropout: Randomly drops 25% of units.
x = Dropout(0.25, name='dropout_3')(x)

# 5. Classifier Head
# Flatten: Converts the 3D feature maps (e.g., 4x4x128) into a 1D vector (2048 units) for the Dense layers.
x = Flatten(name='flatten_layer')(x)
# Dense: A fully connected layer with 512 units to process the flattened features.
x = Dense(512, activation='relu', name='dense_1')(x)
# BatchNormalization: Normalizes activations before the final dropout.
x = BatchNormalization(name='bn_4')(x)
# Dropout: Higher dropout rate (50%) in the classifier head to prevent strong reliance on specific features.
x = Dropout(0.5, name='dropout_4')(x)

# 6. Output Layer
# Dense: The final output layer. It has 'num_classes' (10) units, and the 'softmax' activation
# converts the logits into a probability distribution over the 10 classes.
outputs = Dense(num_classes, activation='softmax', name='output_layer')(x)


# tf.keras.Model: Creates the final model by specifying the input and output layers,
# defining the entire computational graph.
model = tf.keras.Model(inputs=inputs, outputs=outputs, name='cifar10_cnn_model')
```

```python
# TODO: compile your model here
import tensorflow as tf
# Import the base Recall and Precision, which are non-sparse/categorical by default




train_data = train_data.map(one_hot_encode_labels)
test_data = test_data.map(one_hot_encode_labels)

# --- RE-COMPILE YOUR MODEL WITH THE CORRECT LOSS FUNCTION ---
model.compile(
    optimizer='adam',
    # Loss function MUST be CategoricalCrossentropy for one-hot labels
    loss=tf.keras.losses.CategoricalCrossentropy(),
    # Use explicit Categorical metric objects via the full path
    metrics=[
        'accuracy',
        tf.keras.metrics.Recall(name='recall'),
        tf.keras.metrics.Precision(name='precision')
    ]
)
```

```python
# Plot the validation and training data separately
def plot_loss_curves(history):
    """
    Returns separate loss curves for training and validation metrics.
    """
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']

    recall = history.history['recall']
    val_recall = history.history['val_recall']

    precision = history.history['precision']
    val_precision = history.history['val_precision']

    epochs = range(len(history.history['loss']))

    # Plot loss
    plt.plot(epochs, loss, label='training_loss')
    plt.plot(epochs, val_loss, label='val_loss')
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.legend()

    # Plot accuracy
    plt.figure()
    plt.plot(epochs, accuracy, label='training_accuracy')
    plt.plot(epochs, val_accuracy, label='val_accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.legend();

    # Plot recall
    plt.figure()
    plt.plot(epochs, recall, label='training_recall')
    plt.plot(epochs, val_recall, label='val_recall')
    plt.title('Recall')
    plt.xlabel('Epochs')
    plt.legend();

    # Plot precision
    plt.figure()
    plt.plot(epochs, precision, label='training_precision')
    plt.plot(epochs, val_precision, label='val_precision')
    plt.title('Precision')
    plt.xlabel('Epochs')
    plt.legend();
```

```python
# Check out the loss curves
plot_loss_curves(history)
```
